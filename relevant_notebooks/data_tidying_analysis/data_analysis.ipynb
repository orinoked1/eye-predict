# Data Analysis

```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plot
import matplotlib.image as mpimg
```

## Load tidy data into DF

```python
#Read pikle into DF
file_name = 'tidy_data_for_analysis'
df = pd.read_pickle(file_name)
```

## Scale the data points within a specific frame (screen resolution, stimulation resolution, etc')

```python
# select only data points within screen resolution
screenResolutionDf = df[(df['X_axis'] <= 1920) & (df['Y_axis'] <= 1080) 
                       & (df['X_axis'] > 0) & (df['Y_axis'] > 0)] 
```

## Distribution of bids among all subjects trials

```python
screenResolutionDf.groupby(['subjectID', 'trialNum'])['bid'].mean().reset_index().hist(column='bid')
```

## One subjects data analysis

```python
subject101Df = screenResolutionDf[(screenResolutionDf['subjectID'] == 'bmem_short_101')]
subject101Df.groupby(['subjectID', 'trialNum'])['bid'].mean().reset_index().plot(x='trialNum', y='bid', kind='bar',figsize=(17,5))
```

## X_axis points distribution per bid

```python
screenResolutionDf.plot.scatter(x='X_axis', y='bid', figsize=(16,9))
```

```python
bidsDataAbove4 = screenResolutionDf[screenResolutionDf['bid'] >= 4]
bidsDataBelow4 = screenResolutionDf[screenResolutionDf['bid']< 4]
```
